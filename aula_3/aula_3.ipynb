```python
import matplotlib.pyplot as plt 
```

```python
df_maior_duracao_musica = df.groupby('album')['duracao_em_min'].mean().sort_values(ascending = False)
```

```python
df_maior_duracao_musica
```

```python
df_maior_duracao_musica.head(5).plot(kind='bar')
plt.xlabel('Albúns')
plt.ylabel('Média de duração das músicas')
plt.title('Top 5 albúns com maior duração média de música')
plt.show()
```

```python
top_albums = df['album'].value_counts().head(10)
plt.barh(top_albums.index, top_albums.values)
plt.xlabel('Média de Duração')
plt.ylabel('Albúns')
plt.title('Top 10 Albúns Com Mais Música')
plt.show()
```

```python
df_ultima_decada = df[df['release_date'].between(pd.to_datetime('2011'), pd.to_datetime('2020'))]
```

```python
df_por_album = df_ultima_decada.groupby('album')['popularity'].sum().sort_values(ascending=False).head(10)

total_popularidade = df_por_album.sum()
df_porcentagem = df_por_album / total_popularidade * 100 

labels = df_porcentagem.index.tolist()
sizes = df_porcentagem.values.tolist()

cores = ['#1DB954', '#9146FF', '#2D46B9', '#F5B700', '#E22134',  '#2E2E3E', '#3A7DFF', '#BB86FC', '#A8E6CF', '#56C8D8']

figura, grafico = plt.subplots(figsize=(18,6))
grafico.pie(sizes, autopct = '%1.1f%%', colors=cores)
grafico.axis('equal')
plt.title('Popularidade dos albúns do Rolling Stones na última década (top 10 albúns)')
plt.legend(labels, loc='best')

plt.show()
```

```python
import seaborn as sns
```

```python
sns.set(style='whitegrid')
fig, axes = plt.subplots(figsize=(12,6))
sns.boxplot(x='duracao_em_min', data=df)
axes.set_title('Boxplot')
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(12,6))
sns.violinplot(x='duracao_em_min', data=df, color='green')
axes.set_title('Grafico de violino')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

sns.violinplot(x='duracao_em_min', data=df, ax=ax, color='lightgray', zorder=2)
sns.boxplot(x='duracao_em_min', data=df, ax=ax, whis=1.5, color='darkblue', zorder=1)
ax.set_title("Visualização Box Plot e Violin Plot") 
plt.show()
```
